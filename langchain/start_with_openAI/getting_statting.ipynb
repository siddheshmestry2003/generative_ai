```python
# chatprompt Template (it means i tell my llm model you have to behave like this)
from langchain_core.prompts import ChatPromptTemplate
prompt=ChatPromptTemplate.from_messages(
   [
    ("system","You are an ecpert AI engimeer.Provide me answers based on the question"),
    ("user","{input}")
   ]
)
```

```python
'''In LangChain, a chain is simply a pipeline that connects multiple components  
 like prompts, LLMs, and output parsers  so data can flow 
 from one to the next automatically.'''
chain=prompt|llm
responce=chain.invoke({"input":"Can you tell me about Langsmith?"})
print(responce)
```

```python
# out parser :- parser is used to structure the output generate by llm model
# (like in json,numbers,list ,or a custom python object) 

from langchain_core.output_parsers import StrOutputParser
output_parser=StrOutputParser()
chain=prompt|llm|output_parser

respomce=chain.invoke({"input":"can you tell me about langsmith"})
print(responce)
```
