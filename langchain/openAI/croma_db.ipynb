## CROMA DB

#### Chroma is a AI native open source vector database focused on developer productivity and happiness . Chroma is licensed under Apadhe 2.O.


```python
# building simplr vectordb

from langchain_community.document_loaders import TextLoader

from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_chroma import Chroma
```

```python
print("siddhesh")
```

```python
loader=TextLoader("speech.txt")
data=loader.load()
```

```python
data
```

```python
# splitter
text_splitter=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=10)
split=text_splitter.split_documents(data)
```

```python
embedding=OllamaEmbeddings(model="mxbai-embed-large")
vectrodb=Chroma.from_documents(documents=split,embedding=embedding)
```

```python
# quering
query="perfect time"
docs=vectrodb.similarity_search(query)
docs
```

```python
docs=vectrodb.similarity_search_with_score(query)
docs
```

```python
## saving to the disk
vectrodb=Chroma.from_documents(documents=split,embedding=embedding,persist_directory="./chroma_db")
```

```python
# load from disk
db2=Chroma(persist_directory="./chroma_db",embedding_function=embedding)
deos2=db2.similarity_search(query)
deos2[0].page_content
```

```python
# similarity score
does3=db2.similarity_search_with_score(query=query)
does3
```

```python
# Retriever option
retieve=vectrodb.as_retriever()
retieve.invoke(query)
```
