# text Splitter form document

#### this text splitter is the recommended one for generic text itis parameterized by a list of characters.Ti tries to split on then in order until the chunks are small enough the default list is ["\n\n","\n|," ",""]. Thid has the effect of trying to keep all paragraphs(and then sentences and the words ) together as long as possible,as those word generally seem to be the strongerst semantically related pieces of text.

#### how the text is split: by list of characters.
#### how the chunk size is measured: by number of characters

```python
# reading a pdf file
from langchain_community.document_loaders import PyPDFLoader
loader=PyPDFLoader('speech_pdf.pdf')
pdf_doc=loader.load()
pdf_doc
```

### how to recursively split text by character

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter=RecursiveCharacterTextSplitter(chunk_size=200,chunk_overlap=25)
final_documents=text_splitter.split_documents(pdf_doc)
print(final_documents)
```

```python
print(final_documents[0])
print(final_documents[1])
```
