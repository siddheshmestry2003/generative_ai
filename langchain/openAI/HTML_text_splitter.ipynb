# how to split HTML header

```python
from langchain_text_splitters import HTMLHeaderTextSplitter

html_doc = """
<h1>Artificial Intelligence</h1>
<p>AI is the simulation of human intelligence in machines.</p>

<h2>Machine Learning</h2>
<p>Machine learning enables computers to learn from data.</p>

<h3>Supervised Learning</h3>
<p>In supervised learning, models are trained using labeled data.</p>

<h4>Regression</h4>
<p>Regression predicts continuous values, like price or temperature.</p>

<h5>Linear Regression</h5>
<p>Linear regression finds a line that best fits the data.</p>

<h6>Mathematical Formula</h6>
<p>y = mx + b</p>

<h3>Unsupervised Learning</h3>
<p>Unsupervised learning works with unlabeled data to find hidden patterns.</p>

<h2>Deep Learning</h2>
<p>Deep learning uses neural networks with multiple layers.</p>

<h3>Applications</h3>
<ul>
  <li>Computer Vision</li>
  <li>Natural Language Processing</li>
  <li>Speech Recognition</li>
</ul>

<h4>Example Link</h4>
<p>Learn more at <a href="https://example.com">example.com</a>.</p>
"""


header_to_split_on =[
    ("h1",'Header 1'),
    ("h2",'Header 2'),
    ("h3",'Header 3')
]

html_splitter=HTMLHeaderTextSplitter(headers_to_split_on= header_to_split_on)
html_header_splits=html_splitter.split_text(html_doc)
html_header_splits
```

```python
url="https://www.researchgate.net/publication/395862769_Call_for_Papers_-_Special_Issue_in_Phyton_Journal_Genetic_Resources_and_Breeding_Strategies_for_Sustainable_Crop_Yield_Enhancement"
header_to_split_on =[
    ("h1",'Header 1'),
    ("h2",'Header 2'),
    ("h3",'Header 3')
]
html_splitter=HTMLHeaderTextSplitter(headers_to_split_on= header_to_split_on)
html_header_splits=html_splitter.split_text_from_url(url)
html_header_splits
```
