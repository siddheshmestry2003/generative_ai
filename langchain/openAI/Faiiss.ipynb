## Faiss

#### facebook AI Similarity Search (Faiss) is  a library for efficient similarity search and clustering of dense vector . It contain algorithm that search in sets of vectors of an size , up to ones that possibly do not fit in RAM .It  also cintains supporting code for  evaluation and parameter tuning.

```python
from  langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import faiss

from langchain_community.embeddings import OllamaEmbeddings

from langchain_text_splitters import CharacterTextSplitter

loader=TextLoader('speech.txt')
document=loader.load()
text_splitter=CharacterTextSplitter(chunk_size=750,chunk_overlap=30)
docs=text_splitter.split_documents(document)
```

```python
docs
```

```python
from langchain_community.vectorstores import FAISS
embedding=OllamaEmbeddings(model='gemma:2b')
db=FAISS.from_documents(docs,embedding)
db
```

```python
# querry
query="Good morning everyone,"
docs=db.similarity_search(query)
docs[1]
```

## Similarity searc with score

#### there are some FAISS specific methods. one of them is similarity_search_with_score , which allows you to return not only the document but also the distance score of the query to them , the returned distance score is L2 distance,Teherfore a lower score is bettr.

```python
docs_and_score=db.similarity_search_with_score(query)
docs_and_score
```

## As a retriever

###### we can also convert the vectorstore into a retiiver class. this allows usot esily use it  in  order LangChain methods m which largely work with retrievers.

```python
retiriever=db.as_retriever()
docs=retiriever.invoke(query)
docs
```

```python
embedding_vector=embedding.embed_query(query)
embedding_vector
```

```python
docs_score=db.similarity_search_by_vector(embedding_vector)

docs_score
```

```python
## saving and loading 
db.save_local("faiss_index")
```

```python
## load faiss db
new_db=FAISS.load_local("faiss_index",embeddings=embedding,allow_dangerous_deserialization=True)   
docs=new_db.similarity_search(query)
```
