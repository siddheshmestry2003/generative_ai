```python
## text loader  to load .txt file
from langchain_community.document_loaders import TextLoader

loader=TextLoader('speech.txt')# this speech.txt file is converted into document

loader
docs=loader.load()
```

```python
from langchain_text_splitters import CharacterTextSplitter
char_splitter=CharacterTextSplitter(separator="\n\n",chunk_size=200,chunk_overlap=20)
final_char_splitter=char_splitter.split_documents(docs)
final_char_splitter
```
