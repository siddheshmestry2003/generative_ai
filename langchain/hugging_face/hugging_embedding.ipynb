```python
import os
from dotenv import load_dotenv

# Load environment variables from .env
load_dotenv()

# Now retrieve them safely
hf_token = os.getenv("hf_token")
if hf_token is None:
    print("❌ Hugging Face token not found. Check your .env file or path.")
else:
    os.environ["hf_token"] = hf_token
    print("✅ Hugging Face token loaded successfully!")
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
embedding = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
text ="This is atest document."
query_result=embedding.embed_query(text)
query_result
```

```python
len(query_result)
```

```python
doc_result=embedding.embed_documents([text,"This is not a test document"])
doc_result
```
